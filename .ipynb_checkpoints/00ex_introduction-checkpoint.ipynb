1. The MickeyMouse problem

a) Write a program that prints the numbers from 1 to 100. But for multiples of three print Mickey instead of the number and for the multiples of five print Mouse. For numbers which are multiples of both three and five print MickeyMouse

b) Put the result in a tuple and substitute Mickey with Donald and Mouse with Duck

```python
# Exercise 1 
# point a)

def funzione(i):
    topo = ''
    if (i%3 == 0) and (i%5==0):
        topo = 'MickeyMouse'

    elif (i%5 == 0):
        topo = 'Mouse'

    elif (i%3==0):
        topo = 'Mickey'

    if topo:
       print(topo)
    else:
        print(i)


for i in range(1,101):
    funzione(i)
```

```python
# Exercise 1 
# point b)

la_tupla = tuple([funzione(i) for i in range (1,101)])

print(la_tupla)
```

2\. The swap function

Write a function that swap the values of two input variables x and y (whatever the type). Try to do that also without a temporary variable

```python
# Exercise 2

#With temporary variable, (print statements inserted for verification)
def swappo(x,y):
    print(x)
    print(y)
    t = x
    x = y
    y = t
    print(x)
    print(y)
    
swappo('pizza', 5)

#Without temporary variable, (print statements inserted for verification)

def swappo_senzat(x,y):
    print(x)
    print(y)
    x, y = y,x 
    print(x)
    print(y)

swappo_senzat('pizza', 7)
    
```

3\. Computing the distance

Write a function that calculates and returns the euclidean distance between two points *u* and *v*, where *u* and *v* are both 2-tuples *(x,y)*. For example, if *u=(3,0)* and *v=(0,4)*, the function should return 5

```python
#Exercise 3

import math as m

def distance(a,b):
    return m.sqrt(((a[0]-b[0])**2) + ((a[1]-b[1])**2) )


first = input('Insert x coordinate and y coordinate of first point, separated by comma: ')
u = tuple(float(x) for x in first.split(","))
second = input('Insert x coordinate and y coordinate of second point, separated by comma: ')
v = tuple(float(x) for x in second.split(","))

    
distance(u,v)
```

6\. Combination of functions

Write two functions - one that returns the square of a number, and one that returns the cube. Now write a third function that returns the number raised to the 6th power using the two previous functions.

```python
# Exercise 6 

def square(n):
    return n*n

def cube(n):
    return n*n*n

# Technically correct solution
def sixthpow(n):   
    return n*square(n)*cube(n)


# Other solution
def sixthpow1(n):
    return cube(square(n))

```

7\. Cubes

Create a list of the cubes of x for x in *[0, 10]* using:

a) a for loop

b) a list comprehension

```python
#Exercise 7
#Part a)

def cube(n):
    return n*n*n

a = [0]*11
for i in range(11):
    a[i] = cube(i)
    
print(a)
```

```python
#Exercise 7
#Part b)

b = [cube(i) for i in range(11)]

print(b)

```
